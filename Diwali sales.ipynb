# which age group is buying more 

```python
sns.countplot(data=df, x= 'Age Group')
```

```python
plt.figure(figsize=(4,5))
sns.countplot(data=df, x= 'Age Group' , hue='Gender')
```

```python
#total Amount vs Age Group
aplot = sns.countplot(data=df, x= 'Age Group' , hue='Gender')
for bars in aplot.containers:
    aplot.bar_label(bars)
```

```python
sale_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount', data = sale_age)
```

in the above bar graphs age group between 26-35 are more buyers

# in state number of orders 

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.countplot(data=df, x= 'State')
```

```python
# totol number of orders
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sale_state,x='State',y='Orders')
```

```python
# totol number of orders for top 10
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sale_state,x='State',y='Orders')
```

```python
# tatal amount/sales from top 10 states
sale_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sale_state,x='State',y='Amount')
```

from the above bar graphs we can see that most of the order & sales/amount are from Uttar Pradesh,Maharashtre and Kaernataka respectively

# Marital Status purchasing

```python
plt.figure(figsize=(4,3))
sns.countplot(data = df, x='Marital_Status')
```

```python
plt.figure(figsize=(5,5))
aplot =sns.countplot(data = df, x='Marital_Status') 

for bars in aplot.containers:
    aplot.bar_label(bars)
```

```python
df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

```

```python
sale_Status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

plt.figure(figsize=(5,4))
sns.barplot(data = sale_Status, x='Marital_Status',y='Amount',hue='Gender' )
```

from above graphs we can see that most of the buyers are married women and they have high purchasing power 

# Occupation 

```python
sns.set(rc={'figure.figsize':(20,7)})
aplot=sns.countplot(data = df ,x='Occupation')

for bars in aplot.containers:
    aplot.bar_label(bars)
```

```python
sale_Occu = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=sale_Occu,x='Occupation',y='Amount')
```

above graphs we can see most of the buyers are working in IT,  Healthcare and Aviation 

# Product Category

```python
aplot = sns.countplot(data=df,x= 'Product_Category')

for bars in aplot.containers:
    aplot.bar_label(bars)
```

```python
sale_Product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(data=sale_Product,x='Product_Category',y='Amount')
```

above graphs we can see most of the sold Products are from Food , clothing and Electronics category

```python
sale_Product_id = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(data=sale_Product_id,x='Product_ID',y='Orders')
```

```python
# top 10 most sold Products(same thing as above)

fig1, aplot1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

# Conclusion:

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely buy products from Food,clothing and Electronics category
